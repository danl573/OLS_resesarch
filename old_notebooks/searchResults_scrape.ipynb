```python
import json
```

```python
files = ['json.txt','json1.txt','json2.txt','json3.txt','json4.txt']

data = []
for file in files:
    with open(file) as f:
        json_data = json.load(f)
        data.append(json_data)
```

```python
links = []
count = 0
for json_file in data:
    for element in json_file['results']:
        link = element['url']
        print(count, link, end='\n')
        count = count + 1
        links.append(link)
        
print(len(links), "links")
```
