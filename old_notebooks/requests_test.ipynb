```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
```

```python
# Configure the Safari web driver
driver = webdriver.Safari()

# Navigate to the search results page
query = 'ols'
url = f'https://gov.texas.gov/search?q={query}'
driver.get(url)

# Wait for the page to finish loading
wait = WebDriverWait(driver, 20)
# wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, '#search-results')))

# Extract the links from the page
soup = BeautifulSoup(driver.page_source, 'html.parser')
links = soup.find_all('a')

results = soup.find(id="___gcse_0")
news_elements = results.find_all("div", class_="gsc-webResult gsc-result")

for news_element in news_elements:
    link_element = news_element.find("a", class_="href")
    print(link_element)
    print()
```

```python
for news_element in news_elements:
    link_element = news_element.find("a", class_="gs-title")
    print(link_element)
```

```python
for news_element in news_elements:
    print(news_element.prettify(), end="\n\n")
```
